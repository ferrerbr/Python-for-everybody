```python
import numpy as np

    # crea una  matriz apartir de una lista
A = np.array([[5,8,2],[1,5,9]])
print(A)
print("")

    #crea una matriz con todos los elementos en cero
B = np.zeros((2,3),int) # np.zeros((shape),DataType)
print(B)
print("")

    #crea una matriz con todos los elementos en uno
C = np.ones((3,3),int) # np.ones((shape),DataType)
print(C)
print("")

    #crea una matriz con todos los elementos establecidos en un valor en especifico
D = np.full((3,5),5) #np.full((shape),fill_value)
print(D)
print("")

    #crea una matriz(lineal) con valores de un rango dado con tamaño de paso especifico
E = np.arange(5,20,2) #np.arange(start,stop, tamaño del salto)
print(E)
print("")
    #crea una matriz(lineal)  con una secuencia de numeros que estan distribuidos de 
    #forma uniforme en un rango en especifico 
F = np.linspace(0,100,14)   #np.linspace(start,stop,No_elementos)
print(F)
print("")

    #crea una matriz identidad de tamaño N X N 
   #Una matriz identidad  es cuadrada y la diagonal principal  contiene valores de 1
G = np.eye(5) #np.eye(N)     N=tamaño de la matriz
print(G)
print("")

```

```python
import numpy as np

A = np.array([[1,3,4],[5,1,3],[1,2,3]])
B = np.array([[2,6,5,7],[1,4,5,8],[5,5,5,5]])
C = np.array([[2,9,7],[1,4,6],[5,5,5]])

print(f"Matriz A=\n{A}\n\nMatriz B=\n{B} \n\nMatriz C=\n{C}")
print("")


#calcula la media de los elementos de la matriz
media=np.mean(A)
print(media)
print("")

#calcula la mediana de los elementos 
mediana=np.median(A)
print(mediana)
print("")

#calcula la desviacion estandar de los elementos de la matriz
des_estan=np.std(A)
print(des_estan)
print("")

#calcula la varianza de los elementos
varianza=np.var(A)
print(varianza)
print("")

#calcula la suma de los elementos del array
suma=np.sum(A)
print(suma)

```

5_ Indexacion y filtrado

```python

import numpy as np

A = np.array([[1,3,4],[5,1,3],[1,2,3]])
B = np.array([[2,6,5,7],[1,4,5,8],[5,5,5,5]])
C = np.array([[2,9,7],[1,4,6],[5,5,5]])

print(f"Matriz A=\n{A}\n\nMatriz B=\n{B} \n\nMatriz C=\n{C}")
print("")

#accede al elemento de una fila (i) y una columna (j)

elemento=A[1,1]
print(elemento)

#accedd a toda la columna j
columna=A[:,1]
print(columna)

#accedd a toda la fila i
fila=A[1,:]
print(fila)

#filtra la matriz segun una condicion booleana
filtrada =A[A<3]
print(filtrada)


```

6_ Modificacion de matriz y mas 

```python

import numpy as np

A = np.array([[1,3,4],[5,1,3],[1,2,3]])
B = np.array([[2,6,5,7],[1,4,5,8],[5,5,5,5]])
b = np.array([[1],[6],[8]])
C = np.array([[2,9,7],[1,4,6],[5,5,5]])

print(f"Matriz A=\n{A}\n\nMatriz B=\n{B} \n\nMatriz C=\n{C}")
print("")

#modificacion de un elemento de la matriz
A[1,1]=100
print(A)
print("")


#multiplicacion formal de matrices 
D=np.dot(A,B)
print(D)

#inversa de una matriz 
a_inv = np.linalg.inv(A)
print(a_inv)
print("")

#determinante de una matriz
a_det=np.linalg.det(A)
print(a_det)
print("")

#resolver un sistema de ecuaciones 

sol_ecuacion=np.linalg.solve(A,b)
print(sol_ecuacion)



print(np.dot(A,a_inv))
```
