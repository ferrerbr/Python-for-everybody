
Claro, aquí tienes un ejercicio sencillo de SQL en Python utilizando la librería sqlite3. Este ejercicio consistirá en crear una base de datos SQLite, crear una tabla, insertar algunos datos y luego recuperarlos mediante consultas SQL básicas:

```python
#importar la libreria
import sqlite3

#conectarse a la base de datos (si no existe se creara)
conn=sqlite3.connect('ejerciciosqlite.db')

#crear un cursor para ejecutar comandos SQL
cur=conn.cursor()


#crear un tabla llama usuarios
cur.execute('''CREATE TABLE IF NOT EXISTS usuarios(
            id INTEGER PRIMARY KEY,
            nombre TEXTO,
            edad INTEGER
            )''')

#eliminamos los datos si es que ya existen 
cur.execute('DELETE FROM usuarios')


#insertar algunos datos en la tabla 

usuarios=[("juan",25),
          ("maria",30),
          ("pedro",28)]
cur.executemany('INSERT INTO usuarios(nombre,edad)VALUES(?,?)',usuarios)

#guardar los cambios 
conn.commit()

#consultar todos los usuario
cur.execute('SELECT * from usuarios')
print("todos los usuarios")

for dato in cur.fetchall():
    print(dato)

#consultar los usuarios mayores a 27 años
cur.execute('SELECT * from usuarios WHERE edad >27')
print("\nusuarios mayores de 27 años")

for dato in cur.fetchall():
    print(dato)


#cerrar la conexion
conn.close()



```

```python
import sqlite3

# Conexión a la base de datos SQLite (puedes cambiar el nombre de la base de datos según tu caso)
conn = sqlite3.connect('ventas.db')
cursor = conn.cursor()

# Creación de la tabla productos (si no existe)
cursor.execute('''CREATE TABLE IF NOT EXISTS productos (
                    id_producto INTEGER PRIMARY KEY,
                    nombre TEXT,
                    precio REAL
                )''')

# Creación de la tabla ventas (si no existe)
cursor.execute('''CREATE TABLE IF NOT EXISTS ventas (
                    id_venta INTEGER PRIMARY KEY,
                    id_producto INTEGER,
                    cantidad INTEGER,
                    fecha DATE
                )''')

# Insertar algunos datos de ejemplo en la tabla productos
productos_data = [
    (1, 'Producto A', 10.00),
    (2, 'Producto B', 15.00),
    (3, 'Producto C', 20.00),
    (4, 'Producto D', 25.00),
]

cursor.executemany('INSERT INTO productos VALUES (?, ?, ?)', productos_data)
conn.commit()

# Insertar algunos datos de ejemplo en la tabla ventas
ventas_data = [
    (1, 1, 5, '2023-01-15'),
    (2, 2, 3, '2023-02-20'),
    (3, 1, 8, '2023-03-10'),
    (4, 3, 6, '2023-04-05'),
    (5, 4, 2, '2023-01-20'),
    (6, 2, 4, '2023-07-08'),
    (7, 3, 7, '2023-09-12'),
    (8, 1, 3, '2023-10-30'),
    (9, 4, 5, '2023-11-25'),
    (10, 2, 2, '2023-12-05'),
]

cursor.executemany('INSERT INTO ventas VALUES (?, ?, ?, ?)', ventas_data)
conn.commit()

# Consulta SQL para calcular el total de ingresos generados por la venta de cada producto
cursor.execute('''SELECT p.id_producto, p.nombre, SUM(p.precio * v.cantidad) AS total_ingresos
                  FROM productos p
                  INNER JOIN ventas v ON p.id_producto = v.id_producto
                  GROUP BY p.id_producto, p.nombre
                  ORDER BY total_ingresos DESC''')

# Mostrar resultados
resultados = cursor.fetchall()
print("Total de ingresos por producto:")
for producto in resultados:
    print("ID Producto:", producto[0], "- Nombre:", producto[1], "- Total de ingresos:", producto[2])

# Cerrar la conexión
conn.close()
```
