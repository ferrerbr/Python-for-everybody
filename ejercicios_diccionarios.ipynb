Supongamos que queremos crear un programa que gestione una inventario telefónica. Podemos utilizar un diccionario donde cada contacto esté representado por su nombre (clave) y su número de teléfono (valor). El programa debe permitir al usuario agregar nuevos contactos, buscar contactos existentes y eliminar contactos.

```python
agenda=dict()
print(" 1._Agregar nuevo contacto \n 2._Buscar contacto existente \n 3._Eliminar contacto \n 4._Salir")

while True:
    a=input("¿Que desas hacer ?")
    a=a.lower().strip()
    if a == "1"or a =="agregar nuevo contacto":

        print("agregar ")

        nombre = input("Ingresa el nombre de el nuevo contacto").lower().strip()
        if nombre == "cancelar":
            continue

        elif nombre in agenda:
            print("este contacto ya se encuentra en la agenda")

        else:
            while True:
                telefono = input("ingresa el numero telefonico").strip()
                try:
                    telefono=int(telefono)
                    agenda[nombre] = telefono
                    print(f"el contacto {nombre} ha sido agregado con éxito ")
                    break
                except:
                    print("Ingresa solo valores numericos")
 

    elif a =="2" or a == "buscar contacto existente":

        print("buscar")

        nombre = input("Ingresa el nombre del contacto que desas buscar").lower()
        if nombre in agenda:
            numero = agenda[nombre]
            print(f"{nombre}---{numero}")
        else:
            print("el contacto no se encuentra en nuestra base de datos")


        

    elif a == "3" or a == "eliminar contacto":

        print("eliminar")

        nombre = input("Ingresa el nombre del contacto que deseas eliminar ").lower()
        if nombre in agenda:
            del agenda[nombre]
            print(f"El contacto {nombre} ha sido eliminado")
        else:
            print("El contacto no se encuentra en nuestra base de datos, no es posible eliminarlo ")

    elif a == "4" or a == "salir":
        break


```

Supongamos que queremos crear un programa que registre las ventas de una tienda. Podemos utilizar un diccionario donde las claves sean los nombres de los productos y los valores sean las cantidades vendidas de cada producto. El programa debe permitir al usuario agregar nuevas ventas, buscar ventas existentes y calcular el total de ventas

```python
control_ventas=dict()
print(" 1._Nueva venta \n 2._Buscar ventas de un producto \n 3._calcular el total de ventas \n 4._Salir")

while True:
    a=input("¿Que desas hacer ?")
    a=a.lower().strip()
    if a == "1"or a =="nueva venta":

        print("Nueva venta ")

        nombre = input("Ingresa el nombre del producto").lower().strip()
        if nombre == "cancelar":
            continue

        elif nombre in control_ventas:

            control_ventas[nombre]+=1
            print(f"el prodcto {nombre} ha sido vendido  ")

        else:
                
            control_ventas[nombre]=1
            print(f"el prodcto {nombre} ha sido vendido  ")
 

    elif a =="2" or a == "buscar ventas de un producto":

        print("buscar")
        while True:

            nombre = input("Ingresa el nombre del producto que desas buscar").lower().strip()
            if nombre == "cancelar":
                break
            elif nombre in control_ventas:
                numero = control_ventas[nombre]
                print(f"el producto {nombre} ha sidop vendido >>{numero}<< veces")
                break
            else:
                print(f"El producto {nombre} no ha sido encontrado")

    elif a == "3" or a == "calcular el total de ventas":

        total_ventas=list()

        print("calcular")
        for producto in control_ventas:

            total_ventas.append(control_ventas[producto])
        
        ventas=sum(total_ventas)
        print("El total de ventas de todos los producto es ----",ventas)



    elif a == "4" or a == "salir":
        break

```

```python
inventario=dict()
print(" 1._ Agregar nuevo producto \n 2._ Buscar producto existente \n 3._ Actualizar cantidad de producto \n 4._ Eliminar producto \n 5._ Salir")

while True:

    a=input("¿Que desas hacer ?")
    a=a.lower().strip()


    if a == "1" or a =="agregar nuevo producto":

        print("AGREGANDO NUEVO PRODUCTO\n")

        nombre = input("Ingresa el nombre de el nuevo producto").lower().strip()
        if nombre == "cancelar":
            continue

        elif nombre in inventario:
            print(f"El producto {nombre} ya se encuentra en el inventario\n")

        else:
                  
            inventario[nombre] = 1
            print(f"El conproducto {nombre} ha sido agregado con éxito  al inventario\n")
            
 

    elif a =="2" or a == "buscar producto existente":

        print("BUSCANDO PRODUCTO")

        nombre = input("Ingresa el nombre del producto que desas buscar").lower().strip()
        if nombre in inventario:
            numero = inventario[nombre]
            print(f"{nombre}---{numero}\n")
        else:
            print("el producto no se encuentra en nuestra base de datos\n")


 
    elif a == "3" or a == "actualizar producto":

        print("ACTUALIZANDO CANTIDAD DE UN PRODUCTO")
        nombre = input("Ingresa el nombre del productor que deseas actualizar ").lower().strip()

        if nombre in inventario:
            
            cantidad = input("ingresa la cantidad la cantidad de stock disponible")
            while True:
                try:
                    cantidad=int(cantidad)
                    inventario[nombre]=cantidad

                    print(f"El producto {nombre} ha sido actualizado con la cantidad de {cantidad}\n")
                    break
                except:
                    print("ingresa solo datos numericos del tipo entero")
        else:
            print("El producto no se encuentra en nuestra base de datos, no es posible editarlo\n")



    elif a == "4" or a == "eliminar producto":

        print("ELIMINANDO UN PRODUCTO")

        nombre = input("Ingresa el nombre del producto que deseas eliminar ").lower().strip()
        if nombre in inventario:
            del inventario[nombre]
            print(f"El producto {nombre} ha sido eliminado\n")
        else:
            print(f"El producto {nombre} no se encuentra en nuestra base de datos, no es posible eliminarlo\n")
   


    elif a == "5" or a == "salir":

        break
```

El objetivo es implementar este programa utilizando Python y estructuras de datos como listas y diccionarios para gestionar las tareas pendientes.

```python
tareas = dict()

def agregar_tareas(nombre):
    if nombre  in tareas:
        print(f"ya existe un tarea con el nombre ¨{nombre}¨")
    else:
        tareas[nombre]="No completada"
        print(f"La tarea {nombre} ha sido agregada a la lista")


def tarea_completada(nombre):
    if nombre  in tareas:
        tareas[nombre]= "Completada"
        print(f"La tarea {nombre} ha sido marcada como completada")
    else:
        print(f"La tarea {nombre} no se encuentra en la lista")


def mostrar_tareas():
    
    print("\n")

    for tarea,estatus in tareas.items():
        print(f"{tarea}>>>>>{estatus}")
    print("\n")

    
def eliminar_tarea(nombre):
    if nombre in tareas:
        del tareas[nombre]
        print(f"La tarea '{nombre}' ha sido eliminada")
    else:
        print(f"La tarea '{nombre}' no se encuentra en la lista")  

agregar_tareas("hacer la cama")
agregar_tareas("lavar trastes")
agregar_tareas("hacer el super")
mostrar_tareas()
tarea_completada("hacer la cama")
mostrar_tareas()
eliminar_tarea("lavar trastes")
mostrar_tareas()
```
